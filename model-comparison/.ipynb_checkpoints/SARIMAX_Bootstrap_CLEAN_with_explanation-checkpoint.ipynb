# Overview

This notebook presents a concise, end‑to‑end workflow to forecast daily sales using **SARIMAX + Residual Bootstrap**.

**Flow**
1. **Data Preparation**: Load three CSVs (`Total_sales_per_product.csv`, `conversion_rate.csv`, `inventory_summary.csv`) from your `base_path`.  
   Clean, merge, and impute stock‑out periods by weekday means. Produce a daily time series **`combined_final`** with target **`sales`** and save it as `combined_final.csv` back to `base_path`.
2. **Model**: Fit **SARIMAX** on the training set.
3. **Uncertainty**: Generate a **Residual Bootstrap** forecast on the test horizon.
4. **Evaluation**: Print **MAE**, **RMSE**, and **95% CI coverage**; plot **Actual vs Predicted** with CI and on‑plot annotation.
5. *(Optional)* **Forward Forecast**: Produce `HORIZON`‑day forecast with CI bands for presentation.

Everything is written in **English**, with parameters centralized for easy adjustment.



## Parameter Selection Rationale

For this SARIMAX model, we fixed the parameters as follows:

- **Order (p,d,q)** = **(2, 0, 2)**
- **Seasonal Order (P,D,Q,s)** = **(0, 1, 1, 7)**
- **enforce_stationarity** = **True**
- **enforce_invertibility** = **True**

### Why these parameters were chosen

**Data context:** This is a daily sales time series with a clear **weekly pattern** (common in ecommerce and website traffic), and with periods of stock-outs already imputed using weekday averages.

**Non-seasonal order (p,d,q) = (2,0,2):**
- `d=0`: Avoids over-differencing. Data has already been cleaned and weekly adjustments are handled through the seasonal part (`D=1, s=7`). The `enforce_stationarity=True` flag ensures a stable process without unnecessary differencing.
- `p=2`: Captures short-term dependencies (persistence over 1–2 days), often seen in retail sales.
- `q=2`: Captures short-lived shocks or noise (campaigns, events) and corrects autocorrelation in residuals.

**Seasonal order (P,D,Q,s) = (0,1,1,7):**
- `s=7`: Weekly seasonality.
- `D=1`: Removes repeating weekly level shifts (e.g., Monday vs. Saturday patterns), stabilizing the mean without affecting daily variation.
- `Q=1`: Weekly spikes in ACF at lag 7 are captured by a seasonal MA(1) term.
- `P=0`: Keeps the model parsimonious. Seasonal AR terms are only added if PACF shows a significant spike at lag 7.

**Enforcements:**
- `enforce_stationarity=True` and `enforce_invertibility=True` prevent unstable models (roots outside the unit circle), avoiding divergence in bootstrap simulations.

### Next steps if this setup is not optimal:
If residual diagnostics indicate autocorrelation remains, consider:
1. Increasing **Q** or setting **P=1** → `(P,D,Q,7) = (1,1,1,7)`
2. Trying alternative (p,d,q) such as `(1,0,1)` or `(3,0,1)`
3. Using `d=1` only if a slow drift is present in the data

This configuration is a robust starting point for weekly-seasonal ecommerce sales forecasting.


```python
# Imports
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
# Parameters
# Data
base_path = '/Users/anapaulaelizondo/Desktop/Cosmos/data/'  # your Mac path for I/O

# Modeling
ORDER = (2, 0, 2)
SEASONAL_ORDER = (0, 1, 1, 7)
ENFORCE_STATIONARITY = True
ENFORCE_INVERTIBILITY = True

# Bootstrap / Evaluation
TEST_DAYS = 30      # last N days for backtest
HORIZON = 15        # optional forward forecast horizon
N_SIMS = 1000       # number of bootstrap simulations
RANDOM_SEED = 42
CAP_AT_ZERO = False # cap simulated paths at zero
```

```python
# Helper functions
def evaluate_model(y_true, y_pred, name='model'):
    mae = mean_absolute_error(y_true, y_pred)
    rmse = mean_squared_error(y_true, y_pred, squared=False)
    return {'name': name, 'MAE': mae, 'RMSE': rmse}

def fit_sarimax(series, order, seasonal_order, exog=None,
                enforce_stationarity=True, enforce_invertibility=True):
    model = SARIMAX(series, order=order, seasonal_order=seasonal_order,
                    exog=exog,
                    enforce_stationarity=enforce_stationarity,
                    enforce_invertibility=enforce_invertibility)
    return model.fit(disp=False)

def bootstrap_residual_forecast(fitted, horizon, n_sims=500, cap_at_zero=False, random_seed=None):
    rng = np.random.default_rng(random_seed)
    resid = fitted.resid.copy()
    resid = resid[~np.isnan(resid)]
    if resid.size == 0:
        raise ValueError('No residuals available for bootstrap.')
    base = fitted.get_forecast(steps=horizon).predicted_mean.values.astype(float)

    sims = np.zeros((n_sims, horizon), dtype=float)
    for i in range(n_sims):
        noise = rng.choice(resid, size=horizon, replace=True)
        sims[i, :] = base + noise
        if cap_at_zero:
            sims[i, :] = np.maximum(sims[i, :], 0.0)

    mean_fc = sims.mean(axis=0)
    lower = np.percentile(sims, 2.5, axis=0)
    upper = np.percentile(sims, 97.5, axis=0)
    return {'mean_fc': mean_fc, 'lower': lower, 'upper': upper, 'sims': sims}
```

```python
# SECTION 1 — Data preparation: build `combined_final` from CSVs and save it
# Expected CSVs (in base_path):
#   - Total_sales_per_product.csv  (columns include: 'Día', 'Artículos netos vendidos', ...)
#   - conversion_rate.csv          (columns include: 'Día', 'Sesiones', ...)
#   - inventory_summary.csv        (columns include: 'Día', 'Unidades de inventario finales', ...)

# Load raw files
sales_path = os.path.join(base_path, 'Total_sales_per_product.csv')
conv_path  = os.path.join(base_path, 'conversion_rate.csv')
inv_path   = os.path.join(base_path, 'inventory_summary.csv')

sales = pd.read_csv(sales_path)
conversion = pd.read_csv(conv_path)
inventory = pd.read_csv(inv_path)

# Clean sales
sales = sales.rename(columns={'Día': 'day', 'Artículos netos vendidos': 'total_sales'})
sales['day'] = pd.to_datetime(sales['day'], format='%Y-%m-%d', errors='coerce')
sales = sales.dropna(subset=['day']).sort_values('day')

# Clean conversion
conversion = conversion.rename(columns={'Día': 'day', 'Sesiones': 'total_sessions'})
conversion['day'] = pd.to_datetime(conversion['day'], format='%Y-%m-%d', errors='coerce')
conversion = conversion.dropna(subset=['day']).sort_values('day')

# Build combined
combined = pd.merge(sales[['day','total_sales']], conversion[['day','total_sessions']], on='day', how='outer').fillna(0)
combined = (combined.groupby('day', as_index=False).agg({'total_sales': 'sum', 'total_sessions': 'first'}))

# Add inventory (sum by day -> 'inventory')
inventory['day'] = pd.to_datetime(inventory['Día'], format='%Y-%m-%d', errors='coerce')
inv_daily = (inventory.groupby('day', as_index=False)
             .agg({'Unidades de inventario finales': 'sum'})
             .rename(columns={'Unidades de inventario finales': 'inventory'}))

combined_copy = pd.merge(combined, inv_daily, on='day', how='left').fillna({'inventory': 0})
combined_copy = combined_copy.sort_values('day')

# Force inventory = 0 for known stockout dates
zero_dates = [
    "24/02/25","25/02/25","27/02/25","10/03/25","11/03/25","12/03/25","13/03/25","14/03/25","15/03/25",
    "17/03/25","18/03/25","20/03/25","21/03/25","22/03/25","27/03/25","28/03/25","29/03/25","30/03/25","31/03/25",
    "01/04/25","02/04/25","03/04/25"
]
zero_days = pd.to_datetime(zero_dates, dayfirst=True)
combined_copy.loc[combined_copy['day'].isin(zero_days), 'inventory'] = 0

# Impute stockout days (year >= 2025 and inventory <= 0 or NaN) using weekday means
df = combined_copy.copy()
df = df.set_index(pd.to_datetime(df['day'])).sort_index()
df['weekday'] = df.index.weekday

means = df.groupby('weekday')[['total_sales','total_sessions']].mean()

mask = (df.index.year >= 2025) & ((df['inventory'] <= 0) | (df['inventory'].isna()))
df.loc[mask, 'total_sales']    = df.loc[mask, 'weekday'].map(means['total_sales'])
df.loc[mask, 'total_sessions'] = df.loc[mask, 'weekday'].map(means['total_sessions'])

df['ignore'] = False
df.loc[mask, 'ignore'] = True

# Final combined_final
combined_final = df.drop(columns=['weekday'])
combined_final['sales'] = pd.to_numeric(combined_final['total_sales'], errors='coerce')
combined_final = combined_final[['sales','total_sales','total_sessions','inventory','ignore']].dropna(subset=['sales'])

# Save to base_path
output_path = os.path.join(base_path, 'combined_final.csv')
combined_final.reset_index().rename(columns={'index':'day'}).to_csv(output_path, index=False)
print(f'combined_final.csv saved to: {output_path}')
print('Shape:', combined_final.shape)
combined_final.head()
```

```python
# SECTION 2 — SARIMAX + Residual Bootstrap: backtest with KPIs and CI

# Split
assert TEST_DAYS > 0, 'TEST_DAYS must be > 0'
train = combined_final.iloc[:-TEST_DAYS]['sales']
test  = combined_final.iloc[-TEST_DAYS:]['sales']

# Fit
fitted = fit_sarimax(series=train,
                     order=ORDER,
                     seasonal_order=SEASONAL_ORDER,
                     exog=None,
                     enforce_stationarity=ENFORCE_STATIONARITY,
                     enforce_invertibility=ENFORCE_INVERTIBILITY)

# Bootstrap forecast on test window
res_bt = bootstrap_residual_forecast(fitted=fitted,
                                     horizon=len(test),
                                     n_sims=N_SIMS,
                                     cap_at_zero=CAP_AT_ZERO,
                                     random_seed=RANDOM_SEED)

pred_mean = pd.Series(res_bt['mean_fc'], index=test.index)
lower_ci  = pd.Series(res_bt['lower'], index=test.index)
upper_ci  = pd.Series(res_bt['upper'], index=test.index)

# KPIs
metrics = evaluate_model(test.values, pred_mean.values, name='SARIMAX + Residual Bootstrap')
coverage_mask = (test.values >= lower_ci.values) & (test.values <= upper_ci.values)
coverage_pct = coverage_mask.mean() * 100.0
print('Backtest metrics:', {k: (round(v, 3) if isinstance(v, (int, float)) else v) for k, v in metrics.items()})
print(f'95% CI coverage on test set: {coverage_pct:.2f}%')

# Plot
fig, ax = plt.subplots(figsize=(11, 5))
ax.plot(train.index, train.values, label='Train')
ax.plot(test.index,  test.values,  label='Actual (Test)')
ax.plot(test.index,  pred_mean.values, linewidth=2, label='Predicted (mean)')
ax.fill_between(test.index, lower_ci.values, upper_ci.values, alpha=0.2, label='95% CI')
ax.set_title('Backtest: SARIMAX + Residual Bootstrap')
ax.legend()
# On-plot annotation
annot = f"MAE: {metrics['MAE']:.2f} | RMSE: {metrics['RMSE']:.2f} | 95% CI: {coverage_pct:.1f}%"
ax.text(0.01, 0.98, annot, transform=ax.transAxes, va='top', ha='left', fontsize=10,
        bbox=dict(boxstyle='round', alpha=0.15))
plt.tight_layout()
plt.show()
```

```python
# SECTION 3 — (Optional) Forward forecast for HORIZON days
series = combined_final['sales']
fitted_all = fit_sarimax(series=series,
                         order=ORDER,
                         seasonal_order=SEASONAL_ORDER,
                         exog=None,
                         enforce_stationarity=ENFORCE_STATIONARITY,
                         enforce_invertibility=ENFORCE_INVERTIBILITY)

res_fwd = bootstrap_residual_forecast(fitted=fitted_all,
                                      horizon=HORIZON,
                                      n_sims=N_SIMS,
                                      cap_at_zero=CAP_AT_ZERO,
                                      random_seed=RANDOM_SEED)

# Future dates
freq = series.index.inferred_freq or pd.infer_freq(series.index) or 'D'
last_date = series.index[-1]
future_dates = pd.date_range(start=last_date + pd.tseries.frequencies.to_offset(freq),
                             periods=HORIZON, freq=freq)

# Plot
fig, ax = plt.subplots(figsize=(11, 5))
ax.plot(series.index, series.values, label='History')
ax.plot(future_dates, res_fwd['mean_fc'], linewidth=2, label='Forecast (mean)')
ax.fill_between(future_dates, res_fwd['lower'], res_fwd['upper'], alpha=0.2, label='95% CI')
ax.set_title(f'{HORIZON}-day Forecast: SARIMAX + Residual Bootstrap')
ax.legend()
plt.tight_layout()
plt.show()
```
